```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
os.listdir(path)
```

```python
df = pd.read_csv(os.path.join(path, "retail_sales_dataset.csv"))
df.head()
```

```python
df = pd.read_csv(os.path.join(path, "retail_sales_dataset.csv"))
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
cat_sales = df.groupby('Product Category')['Total Amount'].sum()

plt.figure(figsize=(8,5))
cat_sales.plot(kind='bar')
plt.title("Total Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
sales_trend = df.groupby('Date')['Total Amount'].sum()

plt.figure(figsize=(10,5))
sales_trend.plot()
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=20)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Quantity'], df['Total Amount'])
plt.title("Quantity vs Total Amount")
plt.xlabel("Quantity")
plt.ylabel("Total Amount")
plt.show()
```

```python
gender_sales = df.groupby('Gender')['Total Amount'].sum()

plt.figure(figsize=(6,4))
gender_sales.plot(kind='bar')
plt.title("Sales by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Sales")
plt.show()
```
